3 - NORMALLY DISTRIBUTED RANDOM NUMBERS (STD NORMAL)

-- Creating a large sample of normally distributed random numbers

-- Plotting the absolute frequency, relative frequency, PDF, and CDF graphs

START

```python
# Creating a large sample of numbers following a standard normal distribution

std_normal_rand_nos = range_x.standard_normal(size = 10**6)

LL = np.floor(np.min(std_normal_rand_nos))
UL = np.ceil(np.max(std_normal_rand_nos))

if abs(LL) > abs(UL):
    UL = abs(LL)
else:
    LL = -UL

print(std_normal_rand_nos)
print(UL, LL)

bins_A = np.arange(LL, UL+0.01, 0.01)

print(bins_A)
```

```python
# Plotting all the graphs
plt.figure(figsize=(20, 20))

plt.subplot(2,2,1)
plt.hist(x = std_normal_rand_nos, bins = bins_A, color="powderblue")
plt.title("Std_Normal_Dist-Absolute Frequency", fontsize = 20, fontweight = "bold")
plt.xlabel("Bins", fontsize = 15, fontweight = "bold")
plt.ylabel("Absolute Frequency", fontsize = 15, fontweight = "bold")

plt.subplot(2,2,2)
plt.hist(x = std_normal_rand_nos, bins = bins_A, color="powderblue", 
         weights=np.ones(len(std_normal_rand_nos)) / len(std_normal_rand_nos))
plt.title("Std_Normal_Dist-Relative Frequency", fontsize = 20, fontweight = "bold")
plt.xlabel("Bins", fontsize = 15, fontweight = "bold")
plt.ylabel("Relative Frequency", fontsize = 15, fontweight = "bold")

plt.subplot(2,2,3)
plt.hist(x = std_normal_rand_nos, bins = bins_A, color="powderblue", density=True)
plt.title("Std_Normal_Dist-PDF", fontsize = 20, fontweight = "bold")
plt.xlabel("Bins", fontsize = 15, fontweight = "bold")
plt.ylabel("PDF", fontsize = 15, fontweight = "bold")

plt.subplot(2,2,4)
plt.hist(x = std_normal_rand_nos, bins = bins_A, color="powderblue", density=True, cumulative=True)
plt.title("Std_Normal_Dist-CDF", fontsize = 20, fontweight = "bold")
plt.xlabel("Bins", fontsize = 15, fontweight = "bold")
plt.ylabel("CDF", fontsize = 15, fontweight = "bold")

# plt.tight_layout()
plt.show()
```

```python
y_pdf = stats.norm.pdf(x = bins_A, loc = 0, scale = 1)
y_cdf = stats.norm.cdf(x = bins_A, loc = 0, scale = 1)
```

```python
# Plots showing hat the pdf and cdf functions generated by histograms as well as through 
# calculations show more or less convergence

plt.figure(figsize=(20, 20))

plt.subplot(1,2,1)
plt.hist(x = std_normal_rand_nos, bins = bins_A, color="powderblue", density=True)
plt.plot(bins_A, y_pdf, color="red", linewidth=3)
plt.title("Std_Normal_Dist-PDF", fontsize = 20, fontweight = "bold")
plt.xlabel("Bins", fontsize = 15, fontweight = "bold")
plt.ylabel("PDF", fontsize = 15, fontweight = "bold")

plt.subplot(1,2,2)
plt.hist(x = std_normal_rand_nos, bins = bins_A, color="powderblue", density=True, cumulative=True)
plt.plot(bins_A, y_cdf, color="red", linewidth=3)
plt.title("Std_Normal_Dist-CDF", fontsize = 20, fontweight = "bold")
plt.xlabel("Bins", fontsize = 15, fontweight = "bold")
plt.ylabel("CDF", fontsize = 15, fontweight = "bold")

# plt.tight_layout()
plt.show()
```

4 - SHOWCASING THE CENTRAL LIMIT THEOREM

-- If random samples of size 'n' are drawn independently from a population with finite mean Mu and finite variance (sigma)^2, then as n -> inf, the sampling distribution of the sample means approaches a normal distribution with mean Mu and variance (sigma)^2/n, regardless of the shape of the original population distribution.

START

```python
# The CLT will be explained with two different population distributions: Uniform and Normal

# Uniform Distribution

LL = 25
UL = 500

ud_pop_dist = range_x.uniform(low = LL, high = UL, size = 10**6)

ud_pop_mean = np.mean(ud_pop_dist)
ud_pop_std_dev = np.std(ud_pop_dist)

ud_sdsm = []

for i in np.arange(0, 10**3):
    ud_samp_mean = np.mean(range_x.choice(ud_pop_dist, 10**3))
    ud_sdsm.append(ud_samp_mean)

ud_sdsm = np.array(ud_sdsm)
ud_sdsm_mean = np.mean(ud_sdsm)
ud_sdsm_std_dev = np.std(ud_sdsm, ddof=1)

print(
    f"Uniform Distribution Pop-Mean: {ud_pop_mean: 0.4f}", 
    f"Uniform Distribution Pop-Std Dev: {ud_pop_std_dev: 0.4f}",
    f"UD_SDSM-Std Dev (Calc): {ud_pop_std_dev/np.sqrt(10**3): 0.4f}",
    f"UD_SDSM-Mean: {ud_sdsm_mean: 0.4f}",
    f"UD_SDSM-Std Dev: {ud_sdsm_std_dev: 0.4f}",
    sep = "\n"
    )


# Pictorially show that the SDSMs is normally distributed

bin_edges = np.linspace(np.min(ud_sdsm), np.max(ud_sdsm), 30)

ud_sdsm_y = stats.norm.pdf(bin_edges, loc = np.mean(ud_sdsm), scale = np.std(ud_sdsm, ddof=1))

plt.figure(figsize=(20, 20))
plt.hist(x = ud_sdsm, bins = bin_edges, density=True, color="powderblue")
plt.plot(bin_edges, ud_sdsm_y, color = "red")
plt.title("PDF of the Sampling Distribution of Sample Means (Extr from a Uniform Dist Pop)", 
            fontsize = 20, fontweight = "bold")
plt.xlabel("Bin Edges", fontsize = 15, fontweight = "bold")
plt.ylabel("PDF", fontsize = 15, fontweight = "bold")
plt.tight_layout()
plt.show()
```

```python
# Normal Distribution

pop_mean = 45
pop_std_dev = 2

nd_pop_dist = range_x.normal(loc = pop_mean, scale=pop_std_dev, size = 10**6)

nd_pop_mean = np.mean(nd_pop_dist)
nd_pop_std_dev = np.std(nd_pop_dist)

nd_sdsm = []

for i in np.arange(0, 10**3):
    nd_samp_mean = np.mean(range_x.choice(nd_pop_dist, 10**3))
    nd_sdsm.append(nd_samp_mean)

nd_sdsm = np.array(nd_sdsm)
nd_sdsm_mean = np.mean(nd_sdsm)
nd_sdsm_std_dev = np.std(nd_sdsm, ddof=1)

print(
    f"Normal Distribution Pop-Mean: {nd_pop_mean: 0.4f}", 
    f"Normal Distribution Pop-Std Dev: {nd_pop_std_dev: 0.4f}",
    f"ND_SDSM-Std Dev (Calc): {nd_pop_std_dev/np.sqrt(10**3): 0.4f}",
    f"ND_SDSM-Mean: {nd_sdsm_mean: 0.4f}",
    f"ND_SDSM-Std Dev: {nd_sdsm_std_dev: 0.4f}",
    sep = "\n"
    )


# Pictorially show that the SDSMs is normally distributed

bin_edges = np.linspace(np.min(nd_sdsm), np.max(nd_sdsm), 50)

nd_sdsm_y = stats.norm.pdf(bin_edges, loc = np.mean(nd_sdsm), scale = np.std(nd_sdsm, ddof=1))

plt.figure(figsize=(20, 20))
plt.hist(x = nd_sdsm, bins = bin_edges, density=True, color="powderblue")
plt.plot(bin_edges, nd_sdsm_y, color = "red")
plt.title("PDF of the Sampling Distribution of Sample Means (Extr from a Normally Dist Pop)", 
            fontsize = 20, fontweight = "bold")
plt.xlabel("Bin Edges", fontsize = 15, fontweight = "bold")
plt.ylabel("PDF", fontsize = 15, fontweight = "bold")
plt.tight_layout()
plt.show()
```

5 - GENERATING CORRELATED RANDOM NUMBERS (CHOLESKY METHOD)

 - Create two uncorrelated std normal variates
 - Create a third standard normal variate that is correlated to one of the above standard normal variate

 - Z1 and Z2 => Independent Standard Normal Variates
 - Rho (Z1, Z3) = 0.6
 - Z3 = Rho * Z1 + SQRT(1-Rho^2) * Z2
 - Rho (Z2, Z3) = SQRT(1-Rho^2)

START

```python
np.set_printoptions(precision = 4, suppress = True)
```

```python
corr_z1_z3 = 0.6
corr_z2_z3 = np.sqrt(1 - corr_z1_z3**2)
sz = 1000

z1 = range_x.standard_normal(size = sz)
z2 = range_x.standard_normal(size = sz)

z3 = (corr_z1_z3 * z1) + (corr_z2_z3 * z2)

comb_array = np.array([z1, z2, z3])
```

```python
comb_array
```

```python
# To make sure that z1 and z2 are independent (uncorrelated). Their correl coeff must be ~ 0

np.corrcoef(comb_array[0], comb_array[1])
```

```python
dict_x = {
            "mean": [],
            "sigma": [],
            "skewness": [],
            "kurtosis": [],
            "count": [],
            "calc_corr_coeff": []
}

for i in range(0, 3):
    mean_x = round(float(np.mean(comb_array[i])), 4)
    sigma_x = round(float(np.std(comb_array[i], ddof = 0)), 4)
    skew_x = round(float(stats.skew(comb_array[i])), 4)
    kurt_x = round(float(stats.kurtosis(comb_array[i])), 4)
    count_x = round(float(len(comb_array[i])), 4)
    corr_coeff_x = round(float(np.corrcoef(comb_array[i], comb_array[-1])[0, 1]), 4)

    dict_x["mean"].append(mean_x)
    dict_x["sigma"].append(sigma_x)
    dict_x["skewness"].append(skew_x)
    dict_x["kurtosis"].append(kurt_x)
    dict_x["count"].append(count_x)
    dict_x["calc_corr_coeff"].append(corr_coeff_x)


print(dict_x)
```

```python
print(
    "set_corr_coeff_z1,_z3: ", corr_z1_z3, "\n",
    "calc_corr_coeff_z1,_z3: ", dict_x["calc_corr_coeff"][0], "\n",
    "set_corr_coeff_z2,_z3: ", corr_z2_z3, "\n",
    "calc_corr_coeff_z1,_z3: ", dict_x["calc_corr_coeff"][1]
)
```
